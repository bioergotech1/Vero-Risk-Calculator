```python
import pandas as pd
import numpy as np

# Paths
master_path = "C:/Users\HP/OneDrive/Desktop/VERO_code/Phase_1/new_data/codige_master_clean__v2.xlsx"
fi_path = "C:/Users\HP/OneDrive/Desktop/VERO_code/Phase_1/new_data/FI_lab_score.xlsx"

# Load data
master = pd.read_excel(master_path)
fi_raw = pd.read_excel(fi_path)

master.shape, master.columns[:10], fi_raw.head()
```

```python
# Collapse FI to one row per patient
fi = (fi_raw
      .sort_values(["id_paziente"])
      .groupby("id_paziente", as_index=False)
      .first()
)

fi.shape
```

```python
# Merge FI into master by patient_id
fi_ren = fi.rename(columns={"id_paziente": "patient_id"})
master_fi = master.merge(fi_ren, on="patient_id", how="left")

master_fi.shape, master_fi[["patient_id", "FI_LAB", "FI_LAB_score"]].head()
```

```python
# Add FI-derived variables
master_fi["fi_available"] = np.where(master_fi["FI_LAB"].notna(), "Yes", "No")

# Clean FI_LAB_score numeric (may be float because of NaN)
master_fi["FI_LAB_score"] = master_fi["FI_LAB_score"].astype("float")

# Map score to frailty categories
score_to_cat = {
    0: "Low frailty",
    1: "Moderate frailty",
    2: "High frailty"
}
master_fi["fi_lab_category"] = master_fi["FI_LAB_score"].map(score_to_cat)

master_fi[["patient_id", "FI_LAB", "FI_LAB_score", "fi_available", "fi_lab_category"]].head()
```

```python
# Build death model matrix: drop variables per your "remove" list

remove_names_logical = [
    "bmi_value",
    "smoking_status_binary",
    "smoking_years",
    "stadio_TNM",                
    "genotipo_DPYD",             
    "observation_end_reason",
    "Number of Previous Treatment Lines",
    "Other Surgical Intervention",
    "transfusion_received",
    "transfusions_total_n",
    "ricovero_n",                
    "observation_days",           
    "n_unique_active_principles",
    "ordinary_hospitalizations_n",
    "other_comorbidities",
    "comobidity_cat",
    "comorbidity_category_list",
    "number of other pathologies",
    "death_during_observation",
    "adr_left",
    "adr_ctcae_grade",
    "adr_chemo_correlation",
    "side_effect",
    "other",
    "ejection_fraction_percent"
]

# Intersect with actual columns
drop_actual = [c for c in remove_names_logical if c in master_fi.columns]
drop_actual
```

```python
# Construct a simple predictor dictionary: variable, group, note
demographics = [
    "age", "age_group", "gender", "ethnicity", "education_level",
    "employment_status", "bmi_category"
]

lifestyle = [
    "alcohol_consumption", "smoking_status_detail"
]

tumor_diag = [
    "tumor_type", "stadio", "grado_istologico",
    "molecular_alterations", "mutations_present",
    "genotipo_DPYD_type", "tumor_diagnosis_date"
]

treatment = [
    "surgical_intervention", "surgery_type", "prior_surgery",
    "surgery_complications", "oncology_treatment_lines_n",
    "radiotherapy_status", "radiotherapy_start_date",
    "chemio_fine_progressione", "chemio_fine_tossicita",
    "chemio_fine_altro", "treatment_line_n",
    "chemo_schema_name", "chemo_cycles_n", "dose_reduced",
    "n_unique_active_principles", "Hospitalizations Count",
    "oncology_schema_modified"
]

comorbid = [
    "ipertensione", "insufficienza_aortica", "dislipidemie",
    "ipb", "obesita", "cardiopatia_ischemica", "fibrillazione_atriale",
    "bpco", "asma", "diabete_tipo_ii", "reflusso_gastroesofageo",
    "cardiopatia_ipertensiva", "insufficienza_renale",
    "sindrome_depressiva", "anemia", "patologie_psichiatriche",
    "patologie_cardiovascolari", "patologie_gastrointestinali",
    "patologie_cerebrovascolari", "CCI_score"
]

adr_vars = [
    "adr_n_tot", "adr_n_grado1", "adr_n_grado2",
    "adr_n_grado3", "adr_n_grado4", "adr_n_grado5",
    "adr_outcome"
]

labs = [
    "blood_glucose_range", "white_blood_cells_range",
    "red_blood_cells_range", "hemoglobin_range",
    "neutrophils_percent_range", "platelet_count_range",
    "creatinine_range", "ast_got_range", "alt_gpt_range",
    "total_bilirubin_range", "direct_bilirubin_range",
    "albumin_range", "azotemia_range", "gamma_gt_range"
]

frailty = ["FI_LAB", "FI_LAB_score", "fi_lab_category", "fi_available"]

group_map = {}
for v in demographics: group_map[v] = "Demographics & Vitals"
for v in lifestyle: group_map[v] = "Lifestyle"
for v in tumor_diag: group_map[v] = "Tumor & Diagnosis"
for v in treatment: group_map[v] = "Treatment & Care Utilization"
for v in comorbid: group_map[v] = "Comorbidities"
for v in adr_vars: group_map[v] = "ADRs & Outcomes (predictor counts)"
for v in labs: group_map[v] = "Lab Values (Ranges)"
for v in frailty: group_map[v] = "Frailty (FI-Lab)"

dict_rows = []
for col in death_model.columns:
    if col in ["patient_id", "death_outcome", "survival_days",
               "observation_start_date", "observation_end_date", "death_date"]:
        continue  # handled in traceability
    dict_rows.append({
        "variable": col,
        "group": group_map.get(col, "Other / Misc"),
        "description": ""
    })

predictor_dict = pd.DataFrame(dict_rows).sort_values(["group", "variable"])

predictor_dict.head(), predictor_dict["group"].unique()
```

```python
# Define removal lists based on your latest instructions (death-related leakage etc.)
remove_for_all_models = [
    "bmi_value",
    "smoking_status_binary",
    "smoking_years",
    "stadio_TNM",              # tumor_stage_tnm in Italian name
    "genotipo_DPYD",           # dpyd_genotype_known
    "observation_end_reason",
    "Number of Previous Treatment Lines",
    "Other Surgical Intervention",
    "transfusion_received",
    "transfusions_total_n",
    "ricovero_n",              # not present but kept for safety
    "observation_days",        # not present but safety
    "active_principles_n",     # legacy
    "Hospitalizations Count",  # from hospitalization table, not in master
    "other_comorbidities",
    "comobidity_cat",
    "comorbidity_category_list",
    "number of other pathologies",
    "death_during_observation",
    "adr_left",
    "adr_ctcae_grade",
    "adr_chemo_correlation",
    "side_effect",
    "other",
    "ejection_fraction_percent",
]

# For hospitalization and ADR, additionally drop treatment / ADR counts
treatment_followup_sensitive = [
    "surgical_intervention",
    "surgery_type",
    "surgery_complications",
    "oncology_treatment_lines_n",
    "radiotherapy_status",
    "radiotherapy_start_date",
    "ordinary_hospitalizations_n",
    "n_treatment_lines",
    "total_chemo_cycles",
    "received_chemo",
    "treatment_duration_days",
    "any_dose_reduction",
    "any_toxicity",
    "end_due_to_progression",
    "max_combo_regimen_size",
    "total_unique_active_drugs",
    "time_from_diagnosis_to_first_treatment_days",
    "received_cardiotoxic_drug",
    "received_nephrotoxic_drug",
    "received_neurotoxic_drug",
    "received_hematotoxic_drug",
    "received_targeted_therapy",
    "n_unique_active_principles",
    "n_chronic_drugs",
    "polypharmacy_flag",
    "chronic_polypharmacy_flag",
    "n_med_events",
    "medication_duration_days",
    "uses_antithrombotics",
    "uses_antidiabetics_atc",
    "uses_renin_angiotensin_drugs",
    "uses_diuretics",
    "uses_lipid_lowering_drugs",
    "uses_cardiovascular_drugs",
    "uses_metabolic_drugs",
    "uses_respiratory_drugs",
    "uses_cns_drugs",
    "uses_gastrointestinal_drugs",
    "uses_analgesic_drugs",
    "n_cardiovascular_drugs",
    "n_metabolic_drugs",
    "n_respiratory_drugs",
    "n_cns_drugs",
    "n_gastrointestinal_drugs",
    "n_analgesic_drugs",
    "adr_n_tot",
    "adr_n_grado1",
    "adr_n_grado2",
    "adr_n_grado3",
    "adr_n_grado4",
    "adr_n_grado5",
    "end_reason_progression_any_line",
    "chemio_fine_tossicita",
    "end_reason_other_any_line",
]

# Helper to drop missing columns safely
def drop_cols(df, cols):
    cols_existing = [c for c in cols if c in df.columns]
    return df.drop(columns=cols_existing), cols_existing

# Build hospitalization model design matrix
hosp_model_design, dropped_common_hosp = drop_cols(hosp_model, remove_for_all_models)
hosp_model_design, dropped_treat_hosp = drop_cols(hosp_model_design, treatment_followup_sensitive)

# For hospitalization modeling, also remove death outcome and ADR outcome and survival-specific columns
extra_outcome_cols = ["death_outcome", "survival_days", "death_date", "adr_description"]
hosp_model_design, dropped_extra_hosp = drop_cols(hosp_model_design, extra_outcome_cols)

hosp_model_design.shape, dropped_common_hosp[:5], dropped_treat_hosp[:5], dropped_extra_hosp
```

```python
# Severe ADR model design matrix
adr_model_design, dropped_common_adr = drop_cols(adr_model, remove_for_all_models)
adr_model_design, dropped_treat_adr = drop_cols(adr_model_design, treatment_followup_sensitive)
adr_model_design, dropped_extra_adr = drop_cols(adr_model_design, extra_outcome_cols)

adr_model_design.shape
```

```python
# Missingness reports
def build_missingness_report(df, outcome_cols):
    miss = df.isna().mean().reset_index()
    miss.columns = ["variable", "missing_fraction"]
    miss["is_outcome"] = miss["variable"].isin(outcome_cols)
    return miss

hosp_missing = build_missingness_report(
    hosp_model_design,
    outcome_cols=["hospitalization_flag", "hospitalization_event"]
)
adr_missing = build_missingness_report(
    adr_model_design,
    outcome_cols=["severe_adr_flag", "severe_adr_event"]
)

hosp_missing.head(), adr_missing.head()
```

```python
# Build a simple predictors dictionary (same logic for both outcomes)
def build_predictor_dictionary(df, outcome_cols):
    cols = [c for c in df.columns if c not in outcome_cols]
    records = []
    for c in cols:
        if c in ["patient_id", "birth_date", "center_name",
                 "observation_start_date", "observation_end_date",
                 "tumor_diagnosis_date", "Oncology Unit Intake Date"]:
            domain = "Traceability / ID"
        elif c in ["age", "age_group", "gender", "ethnicity", "education_level",
                   "employment_status", "bmi_category"]:
            domain = "Demographics & Lifestyle"
        elif c in ["tumor_type", "tumor_stage_roman", "histological_grade",
                   "molecular_alterations", "mutations_present",
                   "dpyd_genotype_type"]:
            domain = "Tumor & Diagnosis"
        elif c.startswith("n_") or "drug" in c.lower() or "uses_" in c.lower():
            domain = "Medication Burden"
        elif "comorbid" in c.lower() or "hypertension" in c.lower() or "diabetes" in c.lower():
            domain = "Comorbidities"
        elif "range" in c or c.endswith("_range") or "FI_LAB" in c:
            domain = "Labs & Frailty"
        elif c in ["hospitalization_flag", "hospitalization_event",
                   "first_hospitalization_date",
                   "severe_adr_flag", "severe_adr_event",
                   "first_severe_adr_date"]:
            domain = "Outcome / Event"
        else:
            domain = "Other Predictors"
        records.append({
            "variable": c,
            "domain": domain,
            "notes": ""
        })
    return pd.DataFrame(records)

hosp_dict = build_predictor_dictionary(
    hosp_model_design,
    outcome_cols=["hospitalization_flag", "hospitalization_event"]
)
adr_dict = build_predictor_dictionary(
    adr_model_design,
    outcome_cols=["severe_adr_flag", "severe_adr_event"]
)

hosp_dict.head(), adr_dict.head()
```

```python
# Build patient traceability tables (subset of columns)
trace_cols = [
    "patient_id", "birth_date", "age", "age_group", "gender",
    "center_name", "tumor_type", "tumor_diagnosis_date",
    "Oncology Unit Intake Date", "observation_start_date", "observation_end_date"
]

hosp_trace = hosp_model_design[[c for c in trace_cols if c in hosp_model_design.columns]].copy()
adr_trace = adr_model_design[[c for c in trace_cols if c in adr_model_design.columns]].copy()

hosp_trace.head(), adr_trace.head()
```
